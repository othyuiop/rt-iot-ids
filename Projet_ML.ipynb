```python
TARGET = 'Attack_type'
X = df.drop(columns=[TARGET])
y = df[TARGET]
```

```python
#repartition de la target
plt.figure(figsize=(10,6))
sns.countplot(y=y, order=y.value_counts().index)
plt.title("R√©partition des types d'attaques")
plt.show()
```

```python
#3 Preprocessing
#identifier les types de variables
X.dtypes
```

```python

cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(exclude=["object"]).columns

print("Variables cat√©gorielles :", list(cat_cols))
print("Variables num√©riques :", list(num_cols))
```

```python
#encoding
from sklearn.preprocessing import OneHotEncoder

encoder_cat = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

X_cat_encoded = encoder_cat.fit_transform(X[cat_cols])
```

```python
#normaliser la variable
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X[num_cols])
```

```python
#combiner
import numpy as np

X_processed = np.hstack((X_num_scaled, X_cat_encoded))
```

```python
#split train/test 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
#4 feature selection
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=20)

X_train_fs = selector.fit_transform(X_train, y_train)
X_test_fs = selector.transform(X_test)
```

```python
#cleaning feature
from sklearn.feature_selection import VarianceThreshold

# Remove constant features
vt = VarianceThreshold(threshold=0.0)
X_processed = vt.fit_transform(X_processed)
```

```python
#feature selection
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=20)

X_train_fs = selector.fit_transform(X_train, y_train)
X_test_fs = selector.transform(X_test)
```

```python
#importer les modeles 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    AdaBoostClassifier,
    ExtraTreesClassifier
)
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
```

```python
#dictionnaire des modeles
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "ExtraTrees": ExtraTreesClassifier(),
    "SVM": SVC(),
    "NaiveBayes": GaussianNB(),
    "MLP": MLPClassifier(max_iter=500)
}
```

```python
#entrainement et evaluation 
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

results = []

for name, model in models.items():
    model.fit(X_train_fs, y_train)
    y_pred = model.predict(X_test_fs)

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="weighted"),
        "Recall": recall_score(y_test, y_pred, average="weighted"),
        "F1-score": f1_score(y_test, y_pred, average="weighted")
    })

results_df = pd.DataFrame(results)
results_df.sort_values(by="F1-score", ascending=False)
```

```python
##  Tuning des mod√®les candidats
#√Ä l‚Äôissue de la phase de comparaison, les mod√®les AdaBoost et ExtraTrees ont
#obtenu les meilleurs scores F1-score pond√©r√©s. Ces mod√®les ont donc √©t√© retenus
#comme candidats pour l‚Äôoptimisation des hyperparam√®tres.
```

```python
#tunning d'Adaboost
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV

param_grid_ada = {
    "n_estimators": [50, 100, 200],
    "learning_rate": [0.01, 0.1, 1.0]
}

grid_ada = GridSearchCV(
    AdaBoostClassifier(random_state=42),
    param_grid_ada,
    cv=3,
    scoring="f1_weighted",
    n_jobs=-1
)

grid_ada.fit(X_train_fs, y_train)

print("Meilleurs param√®tres AdaBoost :", grid_ada.best_params_)
print("Meilleur F1-score AdaBoost :", grid_ada.best_score_)
```

```python
##tunning d'extratrees
from sklearn.ensemble import ExtraTreesClassifier

param_grid_et = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5]
}

grid_et = GridSearchCV(
    ExtraTreesClassifier(random_state=42),
    param_grid_et,
    cv=3,
    scoring="f1_weighted",
    n_jobs=-1
)

grid_et.fit(X_train_fs, y_train)

print("Meilleurs param√®tres ExtraTrees :", grid_et.best_params_)
print("Meilleur F1-score ExtraTrees :", grid_et.best_score_)

```

```python
##comparaison des 2 modeles 
best_ada = grid_ada.best_estimator_
best_et = grid_et.best_estimator_

from sklearn.metrics import f1_score

y_pred_ada = best_ada.predict(X_test_fs)
y_pred_et = best_et.predict(X_test_fs)

print("F1-score AdaBoost (test) :", f1_score(y_test, y_pred_ada, average="weighted"))
print("F1-score ExtraTrees (test) :", f1_score(y_test, y_pred_et, average="weighted"))
```

```python
# Choix du mod√®le final
#Apr√®s optimisation, le mod√®le ExtraTrees a pr√©sent√© les meilleures performances
#en termes de F1-score pond√©r√© sur le jeu de test. Il a donc √©t√© retenu comme
#mod√®le final pour la phase de d√©ploiement.
best_model = best_et  
```

```python
#evaluation final
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print(classification_report(y_test, best_model.predict(X_test_fs)))

plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test, best_model.predict(X_test_fs)), cmap="Blues")
plt.title("Matrice de confusion - Mod√®le final")
plt.show()
```

```python
#sauvegarde du model final 
import joblib

joblib.dump(best_model, "model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(selector, "selector.pkl")
joblib.dump(encoder_cat, "encoder_cat.pkl")
joblib.dump(cat_cols, "cat_cols.pkl")
joblib.dump(num_cols, "num_cols.pkl")
```
